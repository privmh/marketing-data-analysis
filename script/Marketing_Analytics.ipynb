##Data Cleaning

```python
# Remove extra spaces in column names
df.columns = df.columns.str.strip()
```

```python
# Handle Missing Values
df['Income'] = df['Income'].fillna(df['Income'].median())
```

```python
# Remove Duplicates
df.drop_duplicates(inplace=True)
```

```python
# Convert Date Columns
df['Dt_Customer'] = df['Dt_Customer'].astype('datetime64[ns]')
```

```python
# Standardize Marital Status Column
marital_status_mapping = {
    'Married': 'Married',
    'Together': 'Couple',
    'Single': 'Single',
    'Divorced': 'Divorced',
    'Widow': 'Widow',
    'Alone': 'N/A',
    'YOLO': 'N/A',
    'Absurd': 'N/A'
}

df['Marital_Status'] = df['Marital_Status'].map(marital_status_mapping)
```

```python
# Standardize Education Column

education_mapping = {
    'Basic': 'Low Education',
    '2n Cycle': 'Medium Education',
    'Graduation': 'High Education',
    'Master': 'Very High Education',
    'PhD': 'Very High Education'
}

df['Education'] = df['Education'].map(education_mapping)
```

```python
# Handling Income Outliers

Q1 = df['Income'].quantile(0.25)
Q3 = df['Income'].quantile(0.75)
IQR = Q3 - Q1

Lower_Bound = Q1 - 1.5 * IQR
Upper_Bound = Q3 + 1.5 * IQR

income_mask = (df['Income'] < Lower_Bound) | (df['Income'] > Upper_Bound)

df.loc[income_mask, 'Income'] = df['Income'].median()
```

```python
# Handling Year of Birth Outliers

Q1 = df['Year_Birth'].quantile(0.25)
Q3 = df['Year_Birth'].quantile(0.75)
IQR = Q3 - Q1

Lower_Bound = Q1 - 1.5 * IQR
Upper_Bound = Q3 + 1.5 * IQR

outliers_mask = (df['Year_Birth'] < Lower_Bound) | (df['Year_Birth'] > Upper_Bound)

df.loc[outliers_mask, 'Year_Birth'] = df['Year_Birth'].median()
```

##Feature Engineering

```python
# Calculate customer age
df['Age'] = df['Dt_Customer'].dt.year.max() - df['Year_Birth']
```

```python
# Calculate total number of children and teenagers at home
df['Family_Size'] = df[['Kidhome', 'Teenhome']].sum(axis=1)
```

```python
# Calculate total number of purchases across channels
df['Total_Purchases'] = df[['NumWebPurchases',
                            'NumCatalogPurchases',
                            'NumStorePurchases',
                            'NumDealsPurchases']].sum(axis=1)
```

```python
# Calculate total amount spent across all product categories
df['Total_Spending'] = df[['MntWines',
                           'MntFruits',
                           'MntMeatProducts',
                           'MntFishProducts',
                           'MntSweetProducts',
                           'MntGoldProds']].sum(axis=1)
```

##Data Analysis & Visualization

####Customer Segmentation Analysis


```python
# Segment customers into age groups
def age_group(age):
    if age < 20:
        return 'Teenager'
    elif age < 30:
        return 'Young Adult'
    elif age < 40:
        return 'Adult'
    elif age < 50:
        return 'Middle-Aged Adult'
    elif age < 60:
        return 'Senior Adult'
    else:
        return 'Elderly'

df['Age_Group'] = df['Age'].apply(age_group)

# View age group distribution
df['Age_Group'].value_counts()
```

```python
# Visualize age group distribution
sns.countplot(data=df,
              x='Age_Group',
              edgecolor='black',
              order=df['Age_Group'].value_counts().index)
plt.xticks(rotation=30)
plt.title('Customer Count by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Create income groups using quantiles
q1 = df['Income'].quantile(0.33)
q3 = df['Income'].quantile(0.66)

def income_group(income):
    if income < q1:
        return 'Low Income'
    elif income < q3:
        return 'Medium Income'
    else:
        return 'High Income'

df['Income_Group'] = df['Income'].apply(income_group)

# income group distribution
income_counts = df['Income_Group'].value_counts()
```

```python
# Segment customers by spending behavior based on total spending amount
q1 = df['Total_Spending'].quantile(0.33)
q3 = df['Total_Spending'].quantile(0.66)

def spending_group(value):
    if value <= q1:
        return 'Low Spender'
    elif value <= q3:
        return 'Mid Spender'
    else:
        return 'High Spender'

df['Spending_Segment'] = df['Total_Spending'].apply(spending_group)

# Count customers in each spending segment
spending_counts = df['Spending_Segment'].value_counts()
spending_counts
```

```python
# Visualize spending segmentation
plt.pie(spending_counts,
        labels=spending_counts.index,
        autopct='%1.1f%%',
        startangle=90,
        colors=['#99ff99', '#66b3ff', '#ff9999'])
plt.title('Customer Segmentation by Total Spending')
plt.tight_layout()
plt.show()
```

```python
# Total purchases by channel type
purchase_channels = df[[
    'NumWebPurchases',
    'NumCatalogPurchases',
    'NumStorePurchases']].sum().sort_values(ascending=False)

purchase_channels
```

```python
# Visualize purchase channel distribution
purchase_channels.plot(kind='bar', rot=35, color='lightgreen', edgecolor='black')
plt.title('Total Purchases by Channel')
plt.xlabel('Channel')
plt.ylabel('Total Purchases')
plt.tight_layout()
plt.show()
```

```python
# Segment customers by website engagement (web visits)
q1 = df['NumWebVisitsMonth'].quantile(0.33)
q3 = df['NumWebVisitsMonth'].quantile(0.66)

def engagement_level(visits):
    if visits <= q1:
        return 'Low Engagement'
    elif visits >= q3:
        return 'High Engagement'
    else:
        return 'Medium Engagement'

df['Engagement_Level'] = df['NumWebVisitsMonth'].apply(engagement_level)

# Count customers in each engagement level
df['Engagement_Level'].value_counts()
```

```python
# Visualize engagement levels
sns.countplot(data=df,x='Engagement_Level', color='salmon', edgecolor='black')
plt.title('Customer Engagement by Website Visits')
plt.xlabel('Engagement Level')
plt.ylabel('Count')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
# Segment customers by deal-seeking behavior using median threshold
deal_threshold = df['NumDealsPurchases'].median()

def deal_segment(deals):
    if deals >= deal_threshold:
        return 'Deal Seeker'
    else:
        return 'Non-deal Seeker'

df['Deal_Seeking_Behavior'] = df['NumDealsPurchases'].apply(deal_segment)

# Count customers in each deal-seeking category
df['Deal_Seeking_Behavior'].value_counts()
```

```python
# Visualize deal-seeking behavior
sns.countplot(data=df, x='Deal_Seeking_Behavior', color='skyblue',edgecolor='black')
plt.title('Customer Segmentation by Deal Seeking Behavior')
plt.xlabel('Deal Seeking Behavior')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

####Campaign Performance Analysis


```python
# Summarize the number of customers who accepted each campaign
campaign_data = df[['AcceptedCmp1',
                    'AcceptedCmp2',
                    'AcceptedCmp3',
                    'AcceptedCmp4',
                    'AcceptedCmp5',
                    'Response']].sum().sort_values(ascending=False)

campaign_data
```

```python
# Visualize total accepted campaigns
campaign_data.plot(kind='bar',
                   rot=35,
                   color='steelblue',
                   edgecolor='black')
plt.title('Total Accepted Campaigns')
plt.xlabel('Campaign')
plt.ylabel('Number of Acceptances')
plt.tight_layout()
plt.show()
```

```python
# Calculate the total number of campaigns accepted per customer
df['Total_Accepted'] = df[['AcceptedCmp1',
                           'AcceptedCmp2',
                           'AcceptedCmp3',
                           'AcceptedCmp4',
                           'AcceptedCmp5',
                           'Response']].sum(axis=1)

df['AcceptedAny'] = df['Total_Accepted'] > 0

# Count of customers who accepted any campaign vs none
df['AcceptedAny'].value_counts()
```

```python
# Visualize customer response to campaigns
sns.countplot(data=df, x='AcceptedAny', color='skyblue',edgecolor='black')
plt.title('Customer Response to Campaigns')
plt.xlabel('Accepted Any Campaign')
plt.ylabel('Number of Customers')
plt.xticks([0, 1], ['No', 'Yes'])
plt.tight_layout()
plt.show()
```
